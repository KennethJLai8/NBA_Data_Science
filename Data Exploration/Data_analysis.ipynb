Rookie of The Year Analysis

Players from regular season dataframe are filered to be less than 22 years old to account for rookie players

```python
stats = df_final_roy
stats['YEAR'] = stats['SEASON_ID']
stats = stats.groupby(['Player_ID','SEASON_ID']).sum()
stats = stats.apply(pd.to_numeric)
stats.reset_index(level= ['SEASON_ID'], inplace=True)
stats = stats.apply(pd.to_numeric)
stats.sort_values(by='SEASON_ID', inplace = True)
stats = stats[['SEASON_ID', 'MIN', 'PTS', 'AST', 'REB', 'FGM']]
stats

rok_stats = df_reg
rok_stats = rok_stats[~(rok_stats['Age'] >= 22)] 
rok_stats = rok_stats.groupby(['Player','Year']).mean()
rok_stats.reset_index(level= ['Year'], inplace=True)
rok_stats.sort_values(by='Year', inplace = True)
display(stats)
display(rok_stats)
```
